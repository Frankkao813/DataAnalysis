### Background
In June, the third year junior high school students will take a test called Comprenhensive Program for junior high school students (CAP). Subjects in the test includes the following:

* Chinese 
* English 
* Math 
* Natural Science
* Social Science 
* Chinese composition  

Their performance on tests will be a crucial factor on what senior high school they can go to.

A group of students have been taking courses in a supplementary school, and let's explore their performance together!

The data is collected by google form, and there will be questions that ask them their grade on each subject listed above.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/Frankkao813/DataAnalysis/main/Dataset/CAP_Result.csv', sep = ",")
print(df.info())
```

Below is the structure of the data. 

data explanation: 

1. The second to last column represent gender, score of Chinese, English, Math, Natural Science, Social Science, and Composition of each student.

2. grading scale: 
    * for first to fifth bullet point(Chinese, English, ..., Social Science) is A++, A+, A, B++, B+, B, C respectively

    * For Chinese composition score, as shown in the last column, the grading scale is 0-6.

Below is all the method used to explore the data in the analysis. I use class to wrap all the definition to ensure readability. There will be more detailed explanation about the dataset in the follwoing code and word blocks.

```python
class CAPanalysis:
    def __init__(self, data, nameConversion):
        myMap = {'A++':7, "A+": 6, "A": 5, "B++": 4, "B+": 3, "B": 2, "C":1}
        self.data = data
        self.df_score = self.data.applymap(lambda s: myMap.get(s) if s in myMap else s)
        self.subject_nameConversion = nameConversion
        

    def show_gender(self):
        numGender = self.data['性別'].value_counts().to_dict()
        label = []
        num_people = []
        for key, value in numGender.items():
            label_conversion ={1: "Male", 0: "Female"}
            label.append(label_conversion.get(key,0))
            num_people.append(value)
        plt.title("Gender distribution in the data")
        plt.pie(num_people, labels = label)
        plt.show()  

    def subjectGraph(self, subject_name):
        numCount = self.data[subject_name].value_counts().reindex(['C','B','B+','B++',"A", 'A+','A++'])


        gradeList = [["A++", "A+", "A"], ["B++", "B+", "B"],['C']]
        distributionDict = self.data[subject_name].value_counts().reindex(['C','B','B+','B++',"A", 'A+','A++']).to_dict()
        count = [0,0,0]
        i = 0
        for grades in gradeList:
            for grade in grades:
                letter = grade
                count[i] += distributionDict.get(letter, 0)
            i += 1
        labels = "A", "B", "C"
        
        plt.subplot(1,2,1)
        numCount_graph = numCount.plot.pie(title= self.subject_nameConversion.get(subject_name[0:2],0) + " test score")
        numCount_graph.set_ylabel('percentage of people')
        
        plt.subplot(1,2,2)
        plt.pie(count,labels = labels)
        plt.title("letter grade distribution")
        plt.show()
        
        return count
       

    def composition_score(self):
        comp_score_distribution = self.data["作文成績"].value_counts().reindex([i for i in range(0,7)])
        comp_score_distribution.plot.pie(title = "composition score plot")
        plt.xlabel("score")
        plt.ylabel("number of people")
        plt.show()

    def totalScore(self):
        self.df_score['total'] = self.df_score.iloc[:,2:7].sum(axis=1)
        # show the highest five people
        self.df_score.sort_values("total", ascending=False, inplace= True)
        return self.df_score.head()
    
    def correlation(self, sjt1, sjt2): #subject1, 2
        subject1 = self.df_score[sjt1].tolist()
        subject2 = self.df_score[sjt2].tolist()
        zipped= list(zip(subject1, subject2))
        frequency_dict = {}

        for valuePair in zipped:
            frequency_dict.setdefault(valuePair,0)
            frequency_dict[valuePair] += 1
        valuePair= list(frequency_dict.keys())
        pointSize = np.array(list(frequency_dict.values()))*100

        subject1_new = [val[0] for val in valuePair]
        subject2_new = [val[1] for val in valuePair]
        plt.scatter(subject1_new, subject2_new, pointSize)

        plt.xlabel(self.subject_nameConversion.get(sjt1[0:2],0))
        plt.ylabel(self.subject_nameConversion.get(sjt2[0:2],0))
        plt.title(f"corrlation between {self.subject_nameConversion.get(sjt1[0:2],0)} and {self.subject_nameConversion.get(sjt2[0:2],0)}")
        plt.show()

        
```
