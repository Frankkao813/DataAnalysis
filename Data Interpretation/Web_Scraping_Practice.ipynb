```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
def parse_html(link):
    webpage_response = requests.get(link)
    webpage = webpage_response.content
    soup = BeautifulSoup(webpage, "html.parser")
    return soup
```

```python
prefix = "https://content.codecademy.com/courses/beautifulsoup/"

main_page = parse_html("https://content.codecademy.com/courses/beautifulsoup/shellter.html")
turtle_links = main_page.find_all("a")
links = [prefix+a["href"] for a in turtle_links]
print(links)

heading = []
content = []
for link in links:
    html_code = parse_html(link)
    raw_list = [content.text for content in html_code.findAll("li")]
# regular expression
    pandas_regex = re.compile("(\w*):(.*)")
    column_name = [pandas_regex.search(text).group(1) for text in raw_list]
    row = [pandas_regex.search(text).group(2) for text in raw_list]
    heading.append(column_name)
    content.append(row)

print(content)
print(heading)
turtle_data = pd.DataFrame(content, columns = heading[0])


```

```python
# data cleaning
number_regex = re.compile("\d(\.\d)?")
new_age = [float(number_regex.search(num).group()) for num in turtle_data.AGE]
new_weight = [float(number_regex.search(num).group()) for num in turtle_data.WEIGHT]
turtle_data.AGE = new_age
turtle_data.WEIGHT = new_weight
turtle_data.rename(columns = {"AGE": "AGE (year)", "WEIGHT": "WEIGHT (lbs)"})
    
```
