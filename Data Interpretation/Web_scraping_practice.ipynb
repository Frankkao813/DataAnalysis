### Intro
In this task, I am going to scrape the webpage in the `main_page` variable. According to the owner of website, you can choose what turtle you want to adopt on the website, cool!

There is some links in the webiste, so I will also scrape the content in the link. Let's get started!

### First step
Import the require package, and then wrap the syntax taht is commonly used in webscraping into a function (`parse_html`)

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
def parse_html(link):
    webpage_response = requests.get(link)
    webpage = webpage_response.content
    soup = BeautifulSoup(webpage, "html.parser")
    return soup
```

### Second Step

Let's inspect the html elements in the webpage. We can see that the webpage has a relatively simple structure.

```python
main_page = parse_html("https://content.codecademy.com/courses/beautifulsoup/shellter.html")
print(main_page)
```

### Third Step
Add the prefix and the `href` attribute in the \<a\> tag together to obtain the links. Then, one can use list comprehension to stuff all the links obtained into a single list.

```python
prefix = "https://content.codecademy.com/courses/beautifulsoup/"
turtle_links = main_page.find_all("a")
links = [prefix+a["href"] for a in turtle_links]
print(links)
```

### Fourth Step
Using the link list and the function defined above, we can inspect the html structure in each website. We can agin see that the structure in ecach link is very similar.

```python
for link in links:
    html_code = parse_html(link)
    print(html_code)
```

# Fifth Step
Use dictionary to store the data scraped, and turn it into a pandas dataframe.However, we can see that the table is very messy, so we need to do some cleaning.

```python
    
#Define turtle_data:
turtle_data = {}

#follow each link:
for link in links:
  turtle = parse_html(link)
  turtle_name = turtle.select(".name")[0].get_text()
  stats = turtle.find("ul")
  stats_text = stats.get_text("|")
  turtle_data[turtle_name] = stats_text.split("|")
turtle_df = pd.DataFrame(turtle_data)
```

### Sixth Step

Adopt the following steps to clean the data.
* Drop the row where all of the value is `\n`, as stated in the <font color= "green">#drop na value </font>part
* Next, Define some regular expressions to help us clean the data, and then use the regular expressions
* Finally, Make the `AGE` and `WEIGHT` row to be float to facilitate further computation. I also change the index for `NAME` and `AGE` value to include units for the purpose of decreasing ambuiguity.

```python
import pandas as pd
import re


# drop na value
nan_value = float("NaN")
turtle_df.replace("\n", nan_value, inplace= True)
turtle_df.dropna(subset = ["Aesop"], inplace = True)

# setting row name, define some regular expressions

column_name_regex = re.compile("(\w*):(.*)")
number_regex = re.compile("\d(\.\d)?")
row_name = [column_name_regex.search(element).group(1) for element in list(turtle_df["Aesop"])]


# use regular expressions 
name = list(turtle_df.columns)
for i in range(len(name)):
    turtle_info = [column_name_regex.search(element).group(2) for element in turtle_df.iloc[:,i]]
    turtle_df.iloc[:,i] = turtle_info

# transform the element to list
turtle_df.iloc[0,:] = [float(number_regex.search(element).group()) for element in turtle_df.iloc[0,:].tolist()]
turtle_df.iloc[1, :] = [float(number_regex.search(element).group()) for element in turtle_df.iloc[1,:].tolist()]

row_name_new = ["AGE(year)", "WEIGHT(lbs)", "SEX", "BREED", "SOURCE"]
turtle_df["row"] = row_name_new
turtle_df.set_index("row")

```
